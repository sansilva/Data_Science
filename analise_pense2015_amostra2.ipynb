```
# atribuindo arquivo zip a variável caminho
caminho = "/content/drive/MyDrive/módulo3/PeNSE_2015_AMOSTRA2.zip"

zip = ZipFile(caminho)

caminho_pasta_zipada = zip.namelist()[1]
```

**Então vamos dar uma olhada na célula abaixo, o que é esse nosso arquivo zip que instanciamos. Ele é um objeto daquela pasta zipada no Drive em modo leitura. Podemeos compreender alguns métodos do zipfile. Entre eles, por exemplo, eu tenho o “infolist”.**

```
zip.infolist()
```

**Outro método que eu posso utilizar é o namelist(), podemos chamar o namelist() e fica mais fácil. Ele nos retorna uma lista com os nomes dos arquivos e eu posso pegar o arquivo na posição “1”, no caso, que é justamente o arquivo que eu quero ler.**

```
zip.namelist()[1]
```

**Foi criado uma variável chamada caminho_pasta_zipada. O caminho na pasta zipada eu quero é o zip.namelist( )[1], que é aquele arquivo PENSE_AMOSTRA2_ALUNO.CSV. E é isso que eu preciso passar para o nosso zip.open() que eu estou chamando no nosso método pd.read_csv(). Então passamos para o zip.open() o caminho_pasta_zipada.**

```
dados = pd.read_csv(zip.open(caminho_pasta_zipada), 
                    sep=";",
                    thousands=".",
                    decimal = ",")

dados.head()
```
