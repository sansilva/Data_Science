# **03. Manipulação de Dados**

```
import numpy as np

# .sample roda aleatoriamente
# .seed seta uma número padrão, assim mantemos a amostra com os mesmos dados
np.random.seed(687942)

usaveis.sample(n=7)
```

```
# .sum soma as colnunas
# soma todas as linhas dos Estados (axis=1)

# criando uma coluna Total no DataFrame. DataFrame criar "coluna" igual a alguma coisa
usaveis["Total"] = usaveis.sum(axis=1)
```

```
# .sort_values ordena pela coluna informada, por padrão a ordem de ordenação é crescente. Por isso, usamos ascending=False para ordenar de forma decrescente.
# atribuir a ordenação pela coluna total a variável ordenado_por_total
# .drop("Total", axis=1) excluir a coluna Total após sua ordenação. Observação: drop é a forma contrária do insert utilizado para isnerir a coluna total.
ordenado_por_total = usaveis.sort_values("Total", ascending=False)
ordenado_por_total = ordenado_por_total.drop("Total", axis=1)
ordenado_por_total.head()
```

```
ax = ordenado_por_total.head().T.plot(figsize=(10,6))
# Formatação do eixo y para 02 casas decimais
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```
# .columns[6:] descartnado as 05 primeiras colunas e atribuindo a variável colunas_interessadas
# sobrescrevendo a variável ordenado_por_total
colunas_interessadas = ordenado_por_total.columns[6:]
ordenado_por_total = ordenado_por_total[colunas_interessadas]
```

# **Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado.**

```
# implementando a região de cada Estado na unha mesmo usado o "if". Encontrar um maneira mais fácil.
usaveis.index
regiao = []
for estado in usaveis.index:
  if estado[0] == "1":
    regiao.append("Norte")
  elif estado[0] == "2":
    regiao.append("Nordeste")
  elif estado[0] == "3":
    regiao.append("Sudeste")
  elif estado[0] == "4":
    regiao.append("Sul")
  elif estado[0] == "5":
    regiao.append("Centro-Oeste")

```

```
regiao
```

```
# Criando uma coluna Região no DataFrame com os valores da lista regiao.
usaveis["Região"] = regiao
usaveis["Região"]
```

```
from datetime import date

date.today()

```

```
meses = {
    "Jan": 1,
    "Fev": 2,
    "Mar": 3,
    "Abr": 4,
    "Mai": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Set": 9,
    "Out": 10,
    "Nov": 11,
    "Dez": 12
}

def para_dia(ano_mes: str):
  ano: int = int(ano_mes[:4])
  mes: str = ano_mes[5:]
  mes_numero: int = meses[mes]
  return date(ano, mes_numero, 1)

mensal.index = mensal.index.map(para_dia)

mensal.head()

```

```
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=mensal, x=mensal.index ,y="Bahia")
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Total de gastos - Estado da Bahia", color='b')
plt.ylabel("Gastos")
plt.xlabel("Ano")
plt.xticks(rotation=30)
plt.ylim(0, 350)
#ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=11))
plt.grid(linestyle="--")
plt.show()
```
